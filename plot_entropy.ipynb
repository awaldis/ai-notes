```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.001, 0.999, 200)
log = np.log2(1/x)

fig, ax = plt.subplots(facecolor='lightskyblue',layout="constrained")
fig.suptitle('Components of Information Entropy\n' + \
             r'$H(X) = \sum_{i} p(x_i) \log_2\frac{1}{p(x_i)}$')
ax.set_yticks(np.linspace(0, 10, 11))
ax.set_xticks(np.linspace(0, 1, 11))
ax.plot(x, log, label=r'$\log_2\frac{1}{p(x_i)}$')
ax.plot(x, x, label='$p(x_i)$')
plt.grid(True)
ax.legend()
plt.show()
```

```python
x = np.linspace(0.001, 0.999, 200)
entropy = x * np.log2(1/x) + (1-x) * np.log2(1/(1-x))

fig, ax = plt.subplots(facecolor='lightskyblue',layout="constrained")
fig.suptitle('Information Entropy of a Binary Choice\n' + \
             r'$H(X) = \sum_{i} p(x_i) \log_2\frac{1}{p(x_i)}$')
ax.set_yticks(np.linspace(0, 1, 11))
ax.set_xticks(np.linspace(0, 1, 11))
ax.plot(x, entropy)
plt.grid(True)
plt.show()
```
