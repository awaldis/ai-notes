```python
import matplotlib.pyplot as plt
import numpy as np

# Define the input combinations for A and B
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]

# Define the functions
functions = [
    ("False (0)", lambda A, B: 0),
    ("AND", lambda A, B: A and B),
    ("A AND NOT B", lambda A, B: A and not B),
    ("A", lambda A, B: A),
    ("NOT A AND B", lambda A, B: not A and B),
    ("B", lambda A, B: B),
    ("XOR", lambda A, B: A != B),
    ("OR", lambda A, B: A or B),
    ("NOR", lambda A, B: not (A or B)),
    ("XNOR", lambda A, B: A == B),
    ("NOT B", lambda A, B: not B),
    ("A OR NOT B", lambda A, B: A or not B),
    ("NOT A", lambda A, B: not A),
    ("NOT A OR B", lambda A, B: not A or B),
    ("NAND", lambda A, B: not (A and B)),
    ("True (1)", lambda A, B: 1)
    ]

# Function to plot a truth table
def plot_truth_table(name, func):
    data = [["A", "B", "Output"]]
    for A, B in inputs:
        data.append([A, B, int(func(A, B))])


    # Convert to numpy array for easier plotting
    data = np.array(data)

    fig, ax = plt.subplots()
    ax.axis('tight')
    ax.axis('off')
    table = ax.table(cellText=data, cellLoc='center', loc='center')
    table.set_fontsize(14)
    table.scale(1, 2)

    plt.title(f"Truth table for {name}", fontsize=16, pad=1)
    plt.subplots_adjust(top=0.7)  # Adjust the top of the plot to reduce space
    plt.show()

# Plot each truth table
for name, func in functions:
    plot_truth_table(name, func)
```
